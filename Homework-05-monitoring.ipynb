### Q6. Token stability across runs

```python
import sqlite3
import pandas as pd
import starter

query = "How does the agentic loop keep calling the model until it stops?"
for i in range(4):
    starter.rag.rag(query)
    print(f"completed run {i + 1}")

conn = sqlite3.connect('traces.db')
df = pd.read_sql_query("SELECT rowid, name, input_tokens FROM spans WHERE name='rag.llm' ORDER BY rowid", conn)
print('\nllm input tokens by run:')
print(df.to_string(index=False))
print('\ninput token list:', df['input_tokens'].tolist())
conn.close()
```
